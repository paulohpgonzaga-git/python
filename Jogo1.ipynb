```python
import random

print("|********************|")
print("|Jogo de Adivinhação |")
print("|********************|")

numero_secreto = round(random.randrange(0,101))
print("numero_secreto {}".format(numero_secreto))
total_de_tentativas = 3
rodada = 1
finalizar = "N"

while (finalizar!="S"):
  print("Tentativa {} de {}".format(rodada,total_de_tentativas))
  
  chute_str = input("Digite um valor -> ")
  
  print("Você digitou ",chute_str)
  chute = int(chute_str)

  if (chute < 1 or chute >100):
    print("Digite um valor entre 1 e 100")
    continue

  acertou = numero_secreto==chute
  maior = chute > numero_secreto
  menor = chute < numero_secreto

  if(acertou):
    print("Você acertou")    
  else:
    if(maior):
      print("Você errou! Chute maior que número secreto")
    elif(menor):
      print("Você errou! Chute menor que número secreto")

    
      finalizar = input("Deseja finalizar S/N -> ");
      if (finalizar=="S" or finalizar=="s"):
        print("#Fim do Jogo#")
        break
      else:
        rodada = rodada + 1

```
